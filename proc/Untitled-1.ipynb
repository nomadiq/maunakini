```python
import numpy as np
import matplotlib.pyplot as plt
import os
from subprocess import Popen, PIPE
import random
from collections import OrderedDict 
import maunakini.sched as mksc
```

```python
X = 32
Y = 256
p = 0.2
t1 = 2
t2 = 2
sched = mksc.mobli_schedule(X, Y, t1, t2, p)
plt.figure(figsize=(20,20))
plt.imshow(sched)
plt.show()
```

```python
sched = mksc.hyberts_schedule(X, Y, p)
plt.figure(figsize=(20,20))
plt.imshow(sched)
plt.show()
```

```python
sched = mksc.robson_schedule(X, Y, p, mksc.sine_g, .5, normal=True)
plt.figure(figsize=(20,20))
plt.imshow(sched[0])
plt.show()
```

```python
sched = mksc.robson_schedule(X, Y, p, mksc.sine_g, .5, normal=False)
plt.figure(figsize=(20,20))
plt.imshow(sched[0])
plt.show()
```
